```python
tabla1 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
tabla1.head()
```

```python
tabla1.shape
```

```python
tabla2 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
tabla2.head()
```

```python
tabla2.shape
```

```python
tabla = pd.concat([tabla1, tabla2])
tabla.head()
```

```python
tabla["Number of Open Complaints"] = tabla["Number of Open Complaints"].str.split("/")
```

```python
tabla["Number of Open Complaints"] = tabla["Number of Open Complaints"].apply(lambda x: None if not isinstance(x,list) else x[1])
tabla
```

```python
tabla3 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")
tabla3.head()
```

```python
tabla3.shape
```

```python
tabla3.rename(columns= {'State': 'ST'}, inplace=True)
tabla3.head()
```

```python
tabla3.rename(columns= {'Gender': 'GENDER'}, inplace=True)
tabla3.head()
```

```python
tabla = pd.concat([tabla, tabla3])
tabla
```

```python
tabla = tabla.dropna(how='all')
tabla
```

```python
tabla.reset_index()
```

```python
tabla.columns = tabla.columns.str.replace(" ", "_")
tabla.columns = tabla.columns.str.lower()
tabla
```

```python
tabla.rename(columns= {'st': 'state'}, inplace=True)
tabla
```

```python
tabla.rename(columns= {'st': 'state'}, inplace=True)
tabla
```

```python
tabla['gender'] = tabla["gender"].replace({'Male': 'M', 'Femal' and "female":'F'})
tabla["state"] = tabla["state"].replace({'WA':'Washington', 'Cali':'California', 'AZ':'Arizona'})
tabla["education"] = tabla["education"].replace({'Bachelors':'Bachelor'})
tabla["customer_lifetime_value"] = tabla["customer_lifetime_value"].str.replace('%',' ')
tabla["vehicle_class"] = tabla["vehicle_class"].replace({'Sports Car' and "luxury SUV" and "Luxury Car":'Luxury'})
tabla.head(30)
```

```python
tabla = tabla.fillna("NOT DEFINED")
tabla
```

```python
tabla.drop_duplicates(subset=['customer'])
tabla
```

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*
